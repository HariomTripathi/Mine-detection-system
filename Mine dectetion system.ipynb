```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python
#seperating data and labels
x=sonar_data.drop(columns=60,axis=1) #this stores the data from the sonar_data and drop the 60th coloumn
y=sonar_data[60] #tihs stores the 60th coloumn gform sonar_data
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.1, stratify=y, random_state=2)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

```python
print(x_test)
```

Model Testing ->Logistic Regrasstion Model

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Accuracy of the data 

```python

x_p=model.predict(x_train)
x_ac=accuracy_score(x_p,y_train)
```

```python
print(x_ac)
```

```python
y_p=model.predict(x_test)
y_ac=accuracy_score(y_p,y_test)
```

```python
print(y_ac)
```
